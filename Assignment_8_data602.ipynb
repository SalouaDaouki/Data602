**3. Data Wrangling:**

There is lots of wrangling that is needed to be done for this dataset:

- Renaming some columns so they can be understood,

- Get rid of null values,

- Create a subset of the data to contain only the needed column for further analysis,

- Then, sorting the subset based on the inspection date column.


```python
# Fill missing values with a 0
Inspections_BK_filled = Inspections_BK.fillna(value=0)
Inspections_BK_filled.head()
```

```python
# Rename the columns
dict = {'CAMIS':'Business_ID', 'DBA' : 'Business_Name', 'INSPECTION DATE' : 'INSPECTION_DATE', 'CUISINE DESCRIPTION' : 'CUISINE_DESCRIPTION'}
Inspections_BK_filled.rename(columns=dict, inplace=True)
Inspections_BK_filled.head()
```

```python
# Creating a subset of the dataframe for further analysis
Inspection_BK_df = Inspections_BK_filled[['Business_ID', 'Business_Name', 'CUISINE_DESCRIPTION', 'INSPECTION_DATE', 'ACTION', 'SCORE', 'GRADE', 'GRADE DATE']]
Inspection_BK_df.head()
```

```python
# Sorting the data based on Inspection date column
Inspection_BK_df.sort_values('INSPECTION_DATE')
```

```python
# Selecting the rows for the establishments that are not inspected yet
Inspection_BK_filtered = Inspection_BK_df.loc[Inspection_BK_df['INSPECTION_DATE'] != '01/01/1900'] # 1/1/1900 means that the restaurant has not yet had an inspection because it is new
Inspection_BK_filtered.sort_values('SCORE') # sorting them BY SCORE
```

```python
# Selecting the rows for the establishments that don't have the score 0
Inspection_BK_filtered = Inspection_BK_df.loc[Inspection_BK_df['SCORE'] != 0.0] # 0 as a score means no violation and the grade is usually A
Inspection_BK_filtered.sort_values('SCORE') # sorting them BY SCORE
```

```python
# Grouping the data by cuisine description
grouped_data = Inspection_BK_filtered.groupby('CUISINE_DESCRIPTION')
grouped_data.head()
```

**4. Part I: Visualization using Matplotlib:**



```python
# creating a dataframe with the grades and their totals
grade_sums = Inspections_BK_filled.groupby(['GRADE']).Business_ID.count() #"Business ID" is the unique identifier for each establishment

grade_sums
```

```python
from matplotlib import pyplot as plt
import numpy as np
objects = ('0','A', 'B', 'C', 'N', 'P', 'Z')
y_pos = np.arange(len(objects))
performance = grade_sums

plt.bar(y_pos, performance, align='center', alpha=0.5, color='m', edgecolor='b')
plt.xticks(y_pos, objects)
plt.xlabel('Grade')
plt.ylabel('Number of establishments')
plt.title('What is the distribution of establishemnts grades in Brooklyn')

#adding text labels to each bar
for i, v in enumerate(performance):
  plt.text(i - 0.3, v + 0.2, str(v))


plt.show()

#Source of the code: https://pythonspot.com/matplotlib-bar-chart/
```

According to the DOHMH, "Grade associated with the inspection;

â€¢ N = Not Yet Graded

â€¢ A, B and C = Grade A, B and C

â€¢ Z = Grade Pending

â€¢ P= Grade Pending issued on re-opening following an initial inspection that resulted in a closure"

Based on the bar graph above, many almost 29500 food establishments in Brooklyn has no grade, while around 18400 establishments have no to minimum number of violation (0-13). And only a litlle bit more than 1000 have grade pending (grades p & Z).

Now, let's look the main grades (A, B and C) by cuisine description:

```python
# Let's count the grade by cuisine description
cuisine_grade = Inspections_BK_filled .groupby(['CUISINE_DESCRIPTION', 'GRADE']).agg(count_grade = ('Business_ID', "count")).sort_values(["count_grade"],ascending=False).reset_index()

cuisine_grade.head()
# source of the code: https://pandas.pydata.org/pandas-docs/version/0.22/generated/pandas.core.groupby.DataFrameGroupBy.agg.html
```

```python
# Let's keep only the maim grades (A,B,C)
cuisine_grade_upd = cuisine_grade[cuisine_grade['GRADE'].isin(['A','B','C'])]
# To be able to have a nice bargraph, let's focus on a few cuisines
few_cuisine_grade = cuisine_grade_upd[cuisine_grade_upd["CUISINE_DESCRIPTION"].isin(['American', 'Chinese', 'coffee/Tea', 'Pizza', 'Mediterranean', 'Bakery Products/Desserts'])]
few_cuisine_grade
# source of the code: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.isin.html
```

```python
# pivot the dataframe:
few_cuisine_grade_pivot = few_cuisine_grade.pivot_table(index="CUISINE_DESCRIPTION", columns= "GRADE", values= "count_grade").reset_index()
few_cuisine_grade_pivot
#the source of the code: https://campus.datacamp.com/courses/reshaping-data-with-pandas/introduction-to-data-reshaping?ex=8
```

```python
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# set height of bar
y1 = few_cuisine_grade_pivot['A']
y2 = few_cuisine_grade_pivot['B']
y3 = few_cuisine_grade_pivot['C']

# Set position of bar on X axis
br1 = np.arange(len(y1))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

# Make the plot
plt.bar(br1, y1, color ='r', width = barWidth,
		edgecolor ='grey', label ='Grade A')
plt.bar(br2, y2, color ='g', width = barWidth,
		edgecolor ='grey', label ='Grade B')
plt.bar(br3, y3, color ='b', width = barWidth,
		edgecolor ='grey', label ='Grade C')

# Adding Xticks
plt.xlabel('Cuisine Type', fontweight ='bold', fontsize = 15)
plt.ylabel('Grade count', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(y1))],
		few_cuisine_grade_pivot['CUISINE_DESCRIPTION'])

plt.legend()
plt.show()
# source of the code: https://www.geeksforgeeks.org/bar-plot-in-matplotlib/
```

```python
# Scatter Plot:
plt.scatter(few_cuisine_grade_pivot.A, few_cuisine_grade_pivot.B)
plt.show()
```

```python
# line plot
plt.plot(few_cuisine_grade_pivot.A, few_cuisine_grade_pivot.B)
plt.plot(few_cuisine_grade_pivot.A, few_cuisine_grade_pivot.C)
plt.show()
```

```python
# histogram:
plt.hist(few_cuisine_grade.CUISINE_DESCRIPTION)
plt.xticks(rotation=45)
plt.show()
```
